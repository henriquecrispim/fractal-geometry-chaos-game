```python
# ==============================================================================
# GEOMETRIA COMPUTACIONAL E MODELAGEM DE FRACTAIS
# ==============================================================================

import numpy as np
import matplotlib.pyplot as plt

def gerar_sierpinski(num_pontos=50000):
    """
    Gera as coordenadas do Triângulo de Sierpinski usando o Jogo do Caos.
    """
    # 1. Definir os três vértices de um triângulo equilátero
    vertices = np.array([
        [0.0, 0.0],  # Vértice A (Base Esquerda)
        [1.0, 0.0],  # Vértice B (Base Direita)
        [0.5, np.sqrt(3)/2]  # Vértice C (Topo)
    ])

    # 2. Inicializar a matriz para guardar todos os pontos gerados
    pontos = np.zeros((num_pontos, 2))

    # Começar de um ponto aleatório qualquer dentro do plano
    ponto_atual = np.array([0.5, 0.2])

    print(f"Calculando recursivamente {num_pontos} coordenadas geométricas...")

    # 3. Algoritmo Iterativo do Caos
    for i in range(num_pontos):
        # Escolhe um dos 3 vértices aleatoriamente (0, 1 ou 2)
        vertice_escolhido = vertices[np.random.randint(0, 3)]

        # O próximo ponto será a metade do caminho entre o ponto atual e o vértice escolhido
        ponto_atual = (ponto_atual + vertice_escolhido) / 2.0
        pontos[i] = ponto_atual

    return pontos, vertices

# Executar a modelagem matemática
num_iteraçoes = 100000
pontos, vertices = gerar_sierpinski(num_pontos=num_iteraçoes)

# 4. Renderização Avançada com Matplotlib e Mapa de Cores Dinâmico
plt.figure(figsize=(10, 9))
plt.style.use('dark_background') # Fundo escuro realça as cores do fractal

# Criamos um gradiente de cores baseado na distância dos pontos para o topo para dar efeito 3D
cores = pontos[:, 1]  # Mapeia as cores de acordo com a coordenada Y (altura)

plt.scatter(
    pontos[:, 0], pontos[:, 1],
    s=0.1,                # Tamanho mínimo do ponto para alta resolução
    c=cores,              # Vetor dinâmico de gradiente
    cmap='inferno',       # Mapa de cores quente (preto, vermelho, laranja, amarelo)
    marker='.'
)

# Plotar os 3 vértices principais destacados
plt.scatter(vertices[:, 0], vertices[:, 1], color='cyan', s=30, marker='o', label='Vértices Âncora')

# Configurações estéticas de alta qualidade
plt.title(f'Geometria Fractal: Triângulo de Sierpinski ({num_iteraçoes} Iterações)', fontsize=14, fontweight='bold', pad=15)
plt.axis('off')  # Esconde os eixos para destacar apenas a arte matemática
plt.tight_layout()

# Salvar o artefato visual para o GitHub
plt.savefig('fractal_sierpinski.png', dpi=300, facecolor='black')
plt.show()

print("[SUCESSO] Imagem gráfica 'fractal_sierpinski.png' gerada em alta resolução.")
```
